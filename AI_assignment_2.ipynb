```python
#splitting the data into training and testing sets
X=customers.drop(columns=['Churn'])  #used to train
y= customers['Churn']#for testing
```

```python
#Creating a model
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
# Using XGBClassifier as our model
from xgboost import XGBClassifier
model = XGBClassifier() #  object instantiation
model.fit(x_train,y_train)
y_predict = model.predict(x_test)
```

```python
# here is the accuracy score in decimal, if the accuracy is above 0.75 (75%) it means the model has trained successfuly 
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_predict))
```

```python
# Area Under a graph
from sklearn.metrics import roc_curve, roc_auc_score, roc_auc_score
trp, frp,thresholds = roc_curve(y_test,model.predict_proba(x_test)[:,1])
```
